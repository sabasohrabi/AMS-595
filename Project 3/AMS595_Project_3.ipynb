(10 points) State the differences between assignment, shallow copy, and deep copy. Any reasonable answer will get full credit.

***Answer:***

When we use assignment (for example, x = y), no new object is created. Both variables point to the same memory location. This means that if the object is mutable like a list or dictionary, any change made through one variable will also appear through the other variable. So, assignment only creates another reference to the same object, not a copy. But for a shallow copy. It creates a new outer object but does not create copies of the inner objects stored inside it. It just copies the references to them. To make it clear, if we make a shallow copy of a list of lists, the new list will be independent of the original list itself, but the inner lists are still shared between them. So, modifying the contents of the inner lists in one will also change them in the other. At last, deep copy makes a completely independent copy of the variable, including all objects inside it. It completely copies everything without any relation to the previous variable. Any modification made to one will not affect the other one.

## 2. While Loops

(10 points) Write a Python program that calculates the factorial of a given number (you can assume an positive integer) using a while loop. Display the result to the console.

````python

n = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= n:
    factorial *= i
    i += 1

print(n, " factorial is equal to", factorial)


# test cases
# n = 17, Output: 355687428096000
# n = 13, Output: 6227020800
````

## 3. Functions

(10 points) Write a function that determines if a number is prime.

````python
def is_prime(n):

    if n <= 1:
        return False
    k = 2
    while k * k <= n:
        if n % k == 0:
            return False
        k += 1
    return True

print(is_prime(2))      # Output: True
print(is_prime(10))     # Output: False
print(is_prime(17))     # Output: True
````

## 4. List Comprehension

(10 points) Write a Python function called `prime_factors` that takes a relatively small integer as input and returns a list of its distinct prime factors. The function should use list comprehension to generate the list of distinct prime factors. You may use the function is_prime() defined in Q3.


````python
def prime_factors(n):
  factors = []
  for i in range(2, n + 1):
        if n % i == 0 and is_prime(i):
            factors.append(i)
  return factors
# test cases
print(prime_factors(30))  # Output: [2, 3, 5]
print(prime_factors(56))  # Output: [2, 7]
print(prime_factors(198))  # Output: [2, 3, 11]
print(prime_factors(464))  # Output: [2, 29]
````

## 5. Recursive Functions
(25 points) Write a function that uses recursion to generate the $n$th row of the Pascal's triangle.
```
              1
           1     1
        1     2     1
      1    3     3     1
   1    4     6     4    1
1     5    10    10    5    1
...
...
...
```

````python
def pascal(n):
    if n == 1:
        return [1]
    else:
        prev_row = pascal(n - 1)
        row = [1]
        for i in range(len(prev_row) - 1):
            row.append(prev_row[i] + prev_row[i + 1])
        row.append(1)
        return row

# test case
print(pascal(6))  # output [1, 5, 10, 10, 5, 1]
````

Rewrite the above function using only `while` or `for` loops instead of recursion.
